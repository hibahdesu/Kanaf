```python
import gdown
import zipfile
import os
from PIL import Image  # Importing the Pillow library for image handling

# Direct download link
url = 'https://drive.google.com/uc?id=1xPkNhyhBD-FF6-HwuqH6-D6s-m97zOVn'
output_zip = 'diseases_of_date_palm_leaves.zip'

# Download the ZIP file
gdown.download(url, output_zip, quiet=False)

# Define the extraction directory
extract_folder = 'extracted_data'

# Create the directory if it doesn't exist
os.makedirs(extract_folder, exist_ok=True)

# Extract the ZIP file
with zipfile.ZipFile(output_zip, 'r') as z:
    z.extractall(extract_folder)

# Define the path to the Processed folder
processed_folder_path = os.path.join(
    extract_folder,
    'Diseases of date palm leaves dataset',
    'Infected Date Palm Leaves Dataset',
    'Processed'  # This is the main folder
)

# List of specific subfolders to access
subfolders_to_access = [
    '1. Potassium Deficiency',
    '6. Fusarium Wilt',
    '7. Rachis Blight',
    '5. Leaf Spots',
    '2. Manganese Deficiency',
    '9. Healthy sample',
    '8. Parlatoria Blanchardi',
    '4. Black Scorch',
    '3. Magnesium Deficiency'
]

# Create a dictionary to store images
images = {}

# Iterate through each specified subfolder
for subfolder in subfolders_to_access:
    subfolder_path = os.path.join(processed_folder_path, subfolder)

    # Check if the subfolder exists
    if os.path.isdir(subfolder_path):
        files = os.listdir(subfolder_path)

        for filename in files:
            if filename.endswith('.png'):  # Check for PNG files
                file_path = os.path.join(subfolder_path, filename)

                # Open and process the image
                image = Image.open(file_path)
                images[f"{subfolder}_{filename}"] = image  # Store the image in the dictionary

# Example: Displaying the first image from the dictionary
if images:
    first_image_name = next(iter(images))  # Get the first image key
    first_image = images[first_image_name]
    first_image.show()  # Display the image
    print(f"Displaying image: {first_image_name}")
```

```python
import os
import shutil
import random

# Define source and destination paths
source_folder = '/content/extracted_data/Diseases of date palm leaves dataset/Infected Date Palm Leaves Dataset/Processed'
all_images_folder = '/content/working/All_Renamed_Images'  # Folder to store renamed images with subfolders
train_folder = '/content/working/Train_Image_dataset'      # Train dataset folder
test_folder = '/content/working/Test_Image_dataset'        # Test dataset folder

# Create necessary folders
os.makedirs(all_images_folder, exist_ok=True)
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Define the split ratio
train_ratio = 0.8

# Step 1: Rename images and maintain subfolder structure in 'all_images_folder'
for category in os.listdir(source_folder):
    category_path = os.path.join(source_folder, category)

    # Only proceed if it's a folder (for each disease type)
    if os.path.isdir(category_path):
        # Create the corresponding subfolder in 'all_images_folder'
        subfolder_path = os.path.join(all_images_folder, category.replace(' ', '_'))
        os.makedirs(subfolder_path, exist_ok=True)

        images = os.listdir(category_path)
        for i, image in enumerate(images):
            # Create the new name
            new_name = f"{category.replace(' ', '_')}_{i}.png"

            # Define source and destination paths for renaming
            src_image_path = os.path.join(category_path, image)
            dest_image_path = os.path.join(subfolder_path, new_name)

            # Copy and rename the image to the new subfolder
            shutil.copy(src_image_path, dest_image_path)

# Step 2: Splitting images into train and test datasets
all_images = []
for category in os.listdir(all_images_folder):
    category_path = os.path.join(all_images_folder, category)

    # Collect all images from each subfolder
    if os.path.isdir(category_path):
        images = os.listdir(category_path)
        for image in images:
            all_images.append(os.path.join(category_path, image))

# Shuffle the images for randomness
random.shuffle(all_images)

# Determine the split point
split_index = int(train_ratio * len(all_images))

# Step 3: Split the images into 'train' and 'test' datasets (no subfolders)
for i, image in enumerate(all_images):
    if i < split_index:
        dest_image_path = os.path.join(train_folder, os.path.basename(image))
    else:
        dest_image_path = os.path.join(test_folder, os.path.basename(image))

    # Move the image to the appropriate folder
    shutil.copy(image, dest_image_path)

print("Splitting into training and testing sets completed!")
```

```python
import os
import shutil
import random

# Define source and destination paths
source_folder = '/content/extracted_data/Diseases of date palm leaves dataset/Infected Date Palm Leaves Dataset/Processed'
all_images_folder = '/content/working/All_Renamed_Images'  # Folder to store renamed images with subfolders
train_folder = '/content/working/Train_Image_dataset'      # Train dataset folder
test_folder = '/content/working/Test_Image_dataset'        # Test dataset folder

# Create necessary folders
os.makedirs(all_images_folder, exist_ok=True)
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Define the split ratio
train_ratio = 0.8

# Step 1: Rename images and maintain subfolder structure in 'all_images_folder'
for category in os.listdir(source_folder):
    category_path = os.path.join(source_folder, category)

    # Only proceed if it's a folder (for each disease type)
    if os.path.isdir(category_path):
        # Create the corresponding subfolder in 'all_images_folder'
        subfolder_path = os.path.join(all_images_folder, category.replace(' ', '_'))
        os.makedirs(subfolder_path, exist_ok=True)

        images = os.listdir(category_path)
        for i, image in enumerate(images):
            # Create the new name
            new_name = f"{category.replace(' ', '_')}_{i}.png"

            # Define source and destination paths for renaming
            src_image_path = os.path.join(category_path, image)
            dest_image_path = os.path.join(subfolder_path, new_name)

            # Copy and rename the image to the new subfolder
            shutil.copy(src_image_path, dest_image_path)

# Step 2: Splitting images into train and test datasets with class structure
for category in os.listdir(all_images_folder):
    category_path = os.path.join(all_images_folder, category)

    # Only proceed if it's a directory
    if os.path.isdir(category_path):
        images = os.listdir(category_path)
        random.shuffle(images)  # Shuffle images for randomness

        # Determine the split point
        split_index = int(train_ratio * len(images))

        # Create subfolders in train and test directories
        train_category_path = os.path.join(train_folder, category)
        test_category_path = os.path.join(test_folder, category)
        os.makedirs(train_category_path, exist_ok=True)
        os.makedirs(test_category_path, exist_ok=True)

        # Split the images into 'train' and 'test' datasets
        for i, image in enumerate(images):
            src_image_path = os.path.join(category_path, image)
            if i < split_index:
                dest_image_path = os.path.join(train_category_path, image)
            else:
                dest_image_path = os.path.join(test_category_path, image)

            # Move the image to the appropriate folder
            shutil.copy(src_image_path, dest_image_path)

print("Splitting into training and testing sets completed!")
```

```python
import os
import shutil
import random
import numpy as np
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Define source and destination paths
source_folder = '/content/extracted_data/Diseases of date palm leaves dataset/Infected Date Palm Leaves Dataset/Processed'
train_folder = '/content/working/Train_Image_dataset'
test_folder = '/content/working/Test_Image_dataset'

# Create necessary folders
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# Define the split ratio
train_ratio = 0.8

# Step 1: Rename and organize images
for category in os.listdir(source_folder):
    category_path = os.path.join(source_folder, category)

    if os.path.isdir(category_path):
        # Create subfolders for each class
        train_category_path = os.path.join(train_folder, category.replace(' ', '_'))
        test_category_path = os.path.join(test_folder, category.replace(' ', '_'))
        os.makedirs(train_category_path, exist_ok=True)
        os.makedirs(test_category_path, exist_ok=True)

        images = os.listdir(category_path)
        random.shuffle(images)  # Shuffle images for randomness

        split_index = int(train_ratio * len(images))

        # Split images into training and testing datasets
        for i, image in enumerate(images):
            src_image_path = os.path.join(category_path, image)
            if i < split_index:
                dest_image_path = os.path.join(train_category_path, image)
            else:
                dest_image_path = os.path.join(test_category_path, image)

            shutil.copy(src_image_path, dest_image_path)

print("Data preparation completed!")
```

```python
# Set up the ImageDataGenerator for training and validation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2  # Use 20% for validation
)

# Create training data generator
train_generator = train_datagen.flow_from_directory(
    train_folder,
    target_size=(150, 150),  # Resize images
    batch_size=32,
    class_mode='categorical',  # Adjust based on your number of classes
    subset='training'  # Set as training data
)

# Create validation data generator
validation_generator = train_datagen.flow_from_directory(
    train_folder,
    target_size=(150, 150),
    batch_size=64,
    class_mode='categorical',  # Adjust based on your number of classes
    subset='validation'  # Set as validation data
)
```

```python
# Build the CNN model
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),  # Dropout layer to reduce overfitting
    layers.Dense(train_generator.num_classes, activation='softmax')  # Output layer
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=50,
    callbacks=[early_stopping]
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_folder,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical'
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_generator)
print(f'Test accuracy: {test_accuracy:.2f}')
```
